```python
import torch
from torch import nn

import numpy as np
```

```python
text = ['hey how are you','good i am fine','have a nice day']
```

```python
chars = set(' '.join(text))
chars
```

```python
int2char = dict(enumerate(chars))
int2char
```

```python
char2int = {char: id for id, char in int2char.items()}
char2int
```

```python
maxlen = len(max(text, key=len))
maxlen
```

```python
# Paddings sequences to be equal length
for i in range(len(text)):
  while len(text[i]) != maxlen:
    text[i] += ' '

text
```

```python
# Input data — each character in the sequence (excluding the last one)
# Target data — one time-step ahead of the input char (excluding the first one)
input_seq = []
target_seq = []

for i in range(len(text)):
  input_seq.append(text[i][:-1])

  target_seq.append(text[i][1:])

  print(f'Input sequence: {input_seq[i]}\nTarget sequence: {target_seq[i]}\n')
```

```python
print(input_seq)
print(target_seq)
```

```python
# Transforming into integers
for i in range(len(text)):
  input_seq[i] = [char2int[char] for char in input_seq[i]]
  target_seq[i] = [char2int[char] for char in target_seq[i]]

print(input_seq)
print(target_seq)
```

```python
# number of unique chars
dict_size = len(char2int)
# length of the longest sentence in the dataset minus one because we deleted one char
seq_len = maxlen - 1
# number of elements in dataset (number of sentences)
batch_size = len(text)

def one_hot_encode(sequence, dict_size, seq_len, batch_size):
  features = np.zeros((batch_size, seq_len, dict_size), dtype=np.float32)

  for i in range(batch_size):
    for j in range(seq_len):
      features[i, j, sequence[i][j]] = 1

  return features
```

```python
input_seq = one_hot_encode(input_seq, dict_size, seq_len, batch_size)
```

```python
print(input_seq.shape)
print(input_seq[0].shape)
```

```python
print(type(target_seq))
print(target_seq)

print(type(input_seq))
print(input_seq)
```

```python
# Moving data to torch
input_seq = torch.from_numpy(input_seq)
target_seq = torch.Tensor(target_seq)
```

```python
# Checking if we can use CPU or GPU
is_cuda = torch.cuda.is_available()

if is_cuda:
  device = torch.device('cuda')
  print('GPU is available')
else:
  device = torch.device('cpu')
  print('GPU is not available, CPU is used')
```

# Building a model

```python
class Model(nn.Module):
  def __init__(self, input_size, output_size, hidden_dim, n_layers):
    super(Model, self).__init__()

    # Defining some parameters
    self.hidden_dim = hidden_dim
    self.n_layers = n_layers

    # Defining layers
    # RNN layer
    self.rnn = nn.RNN(input_size, hidden_dim, n_layers, batch_first=True)
    # Fully connected layer
    self.fc = nn.Linear(hidden_dim, output_size)

  def forward(self, x):
    batch_size = x.size(0)  # like .shape in numpy

    # Initializing hidden state for first input using method defined below
    hidden = self.init_hidden(batch_size)

    # Passing in the input and hidden state into the model and obtaining outputs
    output, hidden = self.rnn(x, hidden)

    # Reshaping the outputs such that it can be fit into the fully connected layer
    output = output.contiguous().view(-1, self.hidden_dim)
    output = self.fc(output)

    return output.to(device), hidden.to(device)

  def init_hidden(self, batch_size):
    # This method generates the first initial hidden state of zeros
    hidden = torch.zeros(self.n_layers, batch_size, self.hidden_dim).to(device)
    return hidden
```

```python
model = Model(input_size=dict_size, output_size=dict_size, hidden_dim=12, n_layers=1)
model.to(device)

n_epochs = 100
lr = 0.01

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
```

```python
# Training
for epoch in range(1, n_epochs + 1):
  optimizer.zero_grad()   # to clear existing gradients from previous epoch
  input_seq.to(device)
  output, hidden = model(input_seq)
  loss = criterion(output, target_seq.view(-1).long().to(device))
  loss.backward()   # Does backpropagation and calculates gradients
  optimizer.step() # Updates weights accordingly

  if epoch % 10 == 0:
    print(f'Epoch: {epoch}/{n_epochs}............', end=' ')
    print('Loss: {:.4f}'.format(loss.item()))
    print(output.size())
    print(hidden.size())
    print('\n')
```

```python
def predict(model, character):
  char = np.array([[char2int[c] for c in character]])
  print(char)
  print(char.shape)
  char = one_hot_encode(char, dict_size, char.shape[1], 1)
  print(char)
  char = torch.from_numpy(char)
  print(char)
  char = char.to(device)

  out, hidden = model(char)
  print(f'\n\nout: {out.size()}')
  print(f'\n\nout: {out}')
  print(f'\n\nhidden: {hidden.size()}')

  prob = nn.functional.softmax(out[-1], dim=0).data
  print('\n\nprob: ', prob.size())
  char_ind = torch.max(prob, dim=0)
  print('char_ind:', char_ind)
  char_ind = char_ind[1].item()
  print(char_ind)

  return int2char[char_ind], hidden
```

```python
def sample(model, out_len, start='hey'):
  model.eval()

  start = start.lower()
  chars = [ch for ch in start]
  size = out_len - len(chars)

  for _ in range(size):
    char, h = predict(model, chars)
    chars.append(char)

  return ''.join(chars)
```
